```python
import pandas as pd
```

```python
data = pd.read_csv(r'D:\Academics\MFC Assignment\lending_club_loan_dataset.csv')
data.head(5)
```

```python
# for the sake of simplicity, we are going to drop the categorical data

del data['id']
del data['last_major_derog_none']
del data['term']
del data['grade']
del data["home_ownership"]
del data['purpose']
del data['dti']
```

```python
data.head(5)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.1, random_state= 42)
```

```python
# creating the training labels

y_train = train_set['bad_loan']
train_set.drop('bad_loan', axis=1)
```

```python
# checking the shapes of the train_set and the y_train
print(train_set.shape)
print(y_train.shape)
```

```python
#checking if there's any missing values
train_set.isnull().any()
```

```python
# building the model for predictions

from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Always scale the input. The most convenient way is to use a pipeline.
reg = make_pipeline(StandardScaler(),
                     SGDRegressor(max_iter=1000, tol=1e-3, loss="squared_loss"))

```

```python
reg.fit(train_set, y_train)
```

```python
# predicting on the targets

reg.predict(train_set)
```

```python
# then let's finally checj how the model is scoring in predicting the loan prediction

reg.score(train_set, y_train)
```
